### 6.6. Генераторы  

Генератор - это функция, возвращающая итератор. Эти функции генери­руют объект, который можно перебирать в цикле.

#### 6.6.1. Определение генератора 

> Раз­ница между return и yield заключается в том, что оператор return полностью завер­шает функцию, а оператор yield приостанавливает выполнение функции и сохраня­ет все ее локальные переменные, а затем продолжает выполнение с того же места при последующем вызове или итерации.

````python
# Напишите функцию,которая проверяет,является ли год високосным.


def leap_year_check(year: int) -> None:
    """
    Проверяет, является ли указанный год високосным.

    Параметры:
    year (int): Год, который нужно проверить.

    Возвращает:
    None: Функция ничего не возвращает, но выводит результат
    проверки в консоль.
    """
    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        print(f"{year} is a leap year.")
    else:
        print(f"{year} is not a leap year.")


# Запрос ввода года у пользователя и обработка исключений
while True:
    try:
        user_input = input("Write a year: ")
        num_year = int(user_input)
        leap_year_check(num_year)
        break  # выйти из цикла при успешном вводе
    except ValueError:
        print("Please enter a valid integer for the year.")
````

````python
def convert_cel_to_fahr(celsius: float) -> float:
    """
    #Преобразует температуру из Цельсия в Фаренгейт.

    Параметры:
    celsius (float): Температура в градусах Цельсия.

    Возвращает:
    float: Температура в градусах Фаренгейта.
    """
    return (9 / 5) * celsius + 32


def convert_fahr_to_cel(fahrenheit: float) -> float:
    """
    #Преобразует температуру из Фаренгейта в Цельсий.

    Параметры:
    fahrenheit (float): Температура в градусах Фаренгейта.

    Возвращает:
    float: Температура в градусах Цельсия.
    """
    return (fahrenheit - 32) * 5 / 9


# Ввод температуры в Цельсиях
while True:
    try:
        celsius_temp = float(input("Температура в Цельсиях: "))
        fahrenheit_temp = convert_cel_to_fahr(celsius_temp)
        print(
            f"Температура в Фаренгейтах: {fahrenheit_temp:.2f}"
        )  # Ограничение до двух знаков после запятой
        break  # выйти из цикла после успешного ввода
    except ValueError:
        print("Пожалуйста, введите валидное число.")

# Ввод температуры в Фаренгейтах
while True:
    try:
        fahrenheit_temp_input = float(input("Температура в Фаренгейтах: "))
        celsius_temp_result = convert_fahr_to_cel(fahrenheit_temp_input)
        print(
            f"Температура в Цельсиях: {celsius_temp_result:.2f}"
        )  # Ограничение до двух знаков после запятой
        break  # выйти из цикла после успешного ввода
    except ValueError:
        print("Пожалуйста, введите валидное число.")
````

````python
# Напишите функцию для вычисления факториа лалюбого числа.


def factorial(number: int) -> int:
    """
    Вычисляет факториал заданного целого числа.

    Параметры:
    number (int): Целое число для вычисления факториала.

    Возвращает:
    int: Факториал числа number.

    Исключения:
    ValueError: Если number меньше 0.
    """
    if number < 0:
        raise ValueError(f"Число {number} должно быть не меньше нуля")

    if number in (0, 1):
        return 1

    result = 1
    for i in range(2, number + 1):  # Начинаем с 2, 1 не изменит результат
        result *= i

    return result


# Пример использования
for value in range(11):  # Переменная 'value' для демонстрации
    print(f"Факториал {value} = {factorial(value)}")
````

Напишите функцию, которая принимает в качестве входных данных список и возвращает его перевернутый вариант.

````python
def reverse_list(lst):
  """
  Функция, которая принимает список в качестве входных данных и возвращает его перевернутый вариант.

  Args:
    lst: Список для переворота.

  Returns:
    Перевернутый список.
  """
  return lst[::-1]
```

Пример использования:

```python
my_list = [1, 2, 3, 4, 5]
reversed_list = reverse_list(my_list)
print(reversed_list)  # Вывод: [5, 4, 3, 2, 1]
````

````python
def compound_interest(principal, rate, time, n):
  """
 Функция вычисляет сложные проценты.

  Args:
    principal: Начальная сумма денег.
    rate: Годовая процентная ставка в десятичной форме.
    time: Время в годах.
    n: Число раз в год, когда проценты капитализируются.

  Returns:
    Сумма сложных процентов.
  """

  amount = principal  (1 + (rate / n))(n  time)
  return amount

# Пример использования функции
principal = 10000
rate = 0.05
time = 5
n = 12

compound_interest_amount = compound_interest(principal, rate, time, n)
print("Сложные проценты: ", compound_interest_amount)
````

````python
# Напишите функцию, которая проверяет, является ли переданное число про­стым.


def is_prime(number: int) -> bool:
    """
    Проверяет, является ли заданное целое число простым.

    Должно быть больше 1, делиться только на 1 и само себя.

    Параметры:
    number (int): Целое число, которое необходимо проверить.

    Возвращает:
    bool: True, если число простое, иначе False.
    """
    if number <= 1:
        return False

    for i in range(
        2, int(number**0.5) + 1
    ):  # Проверяем делители до квадратного корня числа
        if number % i == 0:
            return False
    return True


# Пример использования
for num in range(1, 30):
    print(f"{num} {'простое' if is_prime(num) else 'не простое'} число")
````

````python
# Напишите функцию преобразования любого числа от 1 до 1 00 в римское число.


all_roman = [
    (1000, "M"),
    (900, "CM"),
    (500, "D"),
    (400, "CD"),
    (100, "C"),
    (90, "XC"),
    (50, "L"),
    (40, "XL"),
    (10, "X"),
    (9, "IX"),
    (5, "V"),
    (4, "IV"),
    (1, "I"),
]


def conv_to_roman(number: int) -> str:
    """
    Преобразует целое число в римское число.

    Параметры:
    number (int): Целое число от 1 до 3999, которое требуется преобразовать.

    Возвращает:
    str: Римское представление заданного числа.
    """
    roman = ""
    while number > 0:
        for _value, symbol in all_roman:
            while number >= _value:
                roman += symbol
                number -= _value
    return roman


# Пример использования
print(f"Число 2022 в римской системе = {conv_to_roman(2022)}")
````

Напишите функцию f ( х ) , где х - любое 6-значное число, а функция возвращает сумму его цифр. x - num

````python
def digit_sum(number: int) -> int:
    """
    Вычисляет сумму цифр заданного целого числа.

    Параметры:
    number (int): Целое число, сумма цифр которого будет вычислена.

    Возвращает:
    int: Сумма цифр указанного числа.
    """
    if number == 0:
        return number
    sum_ = 0
    while number:
        digit, number = number % 10, number // 10
        sum_ += digit
    return sum_


# Пример использования
print(digit_sum(123456))  # Вывод: 21
````

Напишите функцию f (х) , возвращающую простые множители любого числа х (пример простых множителей: 36- [2 2 3 3], 30- [2 3 5]). x - num

````python
def prime_factors(num):
  """Возвращает список простых множителей заданного числа.

  Args:
    num: Целое число.

  Returns:
    Список простых множителей.
  """

  factors = []
  i = 2
  while i  i <= num:
    while num % i == 0:
      factors.append(i)
      num //= i
    i += 1
  if num > 1:
    factors.append(num)
  return factors

# Пример использования
number = 36
print(f"Простые множители {number}: {prime_factors(number)}")

number = 30
print(f"Простые множители {number}: {prime_factors(number)}")
````
